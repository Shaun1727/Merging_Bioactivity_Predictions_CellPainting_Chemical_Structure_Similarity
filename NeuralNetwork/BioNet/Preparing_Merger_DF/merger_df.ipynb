```python
import pandas as pd

fp_pred_df = pd.read_csv("../../Datasets/finger_print_predictions.csv").drop(["Unnamed: 0"], axis = 1)
cp_pred_df = pd.read_csv("../../Datasets/cell_painting_predictions.csv").drop(["Unnamed: 0"], axis = 1)

merged_df = pd.merge(fp_pred_df,cp_pred_df,on=["assay","index"],suffixes=('_fp', '_cp'))
merged_df = merged_df.rename(columns={'assay_fp': 'assay'})
merged_df.to_csv("../../Datasets/merger_model_df.csv")
```
